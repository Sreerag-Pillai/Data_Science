# Plotting Tesla Historic OHLC Graph

```python
# Importing necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from mplfinance.original_flavor import candlestick_ohlc
```

```python
# Getting stock data from the yf
historic_data = yf.Ticker("TSLA").history(start="2010-01-01", end="2023-09-02")
historic_data.head()
```

Plotting the OHCL bar graph

```python

# Converting the date index to numerical format for matplotlib plotting
historic_data['Date_Num'] = mdates.date2num(historic_data.index.to_pydatetime())

# Create a plot for OHLC bar graph
fig, ax = plt.subplots(figsize=(14, 6))

# Loop through and plot OHLC bars
for date, row in historic_data.iterrows():
    open_price = row['Open']
    high_price = row['High']
    low_price = row['Low']
    close_price = row['Close']
    date_num = row['Date_Num']

    color = 'g' if close_price >= open_price else 'r'

    ax.plot([date_num, date_num], [low_price, high_price], color=color, linewidth=0.5)
    ax.plot([date_num, date_num - 0.1], [open_price, open_price], color=color, linewidth=0.5)
    ax.plot([date_num, date_num + 0.1], [close_price, close_price], color=color, linewidth=0.5)

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.set_title('Tesla OHLC Bar Graph')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

plt.tight_layout()
plt.show()
```

Plotting OHLC candlestick graph

```python
# Create plot for the candlestick graph
fig, ax = plt.subplots(figsize=(14, 6))

candlestick_ohlc(ax, historic_data[['Date_Num', 'Open', 'High', 'Low', 'Close']].values, width=0.6, colorup='g', colordown='r')

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.set_title('Tesla Candlestick Graph')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

plt.tight_layout()
plt.show()
```
